```
import matplotlib.pyplot as plt
data = {
    'Borough': ['Manhattan', 'Bronx', 'Brooklyn', 'Queens', 'Staten Island'],
    'Accidents': [5234, 2345, 3456, 4567, 678]
}
# Create a DataFrame from the data
borough_by_accidents = pd.DataFrame(data)

# Set the 'Borough' column as the index of the DataFrame
borough_by_accidents.set_index('Borough', inplace=True)

# Plot a horizontal bar chart of the DataFrame
ax = borough_by_accidents.plot(kind='barh', legend=False)

# Add labels to each bar with the number of accidents
for i in ax.containers:
    ax.bar_label(i, label_type='edge')

# Set the label for the x-axis
plt.xlabel('Number of Accidents')

# Set the label for the y-axis
plt.ylabel('Borough')
plt.title('Number of Accidents by Borough')
plt.show()

```

```

# Creating the DataFrame
data = {
    'Street': [
        'BROADWAY', 'ATLANTIC AVENUE', 'BELT PARKWAY', '3 AVENUE',
        'NORTHERN BOULEVARD', 'LONG ISLAND EXPRESSWAY', 'BROOKLYN QUEENS EXPRESSWAY',
        'FLATBUSH AVENUE', 'LINDEN BOULEVARD', 'QUEENS BOULEVARD'
    ],
    'Accidents': [
        17277, 15322, 13543, 12476, 11965, 9928, 9743, 9741, 9587, 9368
    ]
}

# Creating the DataFrame
High_accidentials_streets = pd.DataFrame(data)
High_accidentials_streets.set_index('Street', inplace=True)


# Plotting the pie chart
ax = High_accidentials_streets.plot(kind='pie', y='Accidents', autopct='%1.1f%%')

# Adding a legend
plt.legend(
    labels=High_accidentials_streets.index,
    title="Streets",
    loc="center left",
    bbox_to_anchor=(1.5, 0.5)
)

# Customizing the plot
plt.ylabel('') # Remove the default y-axis label
plt.title('Accidents by Street')
plt.show()
```

```
import folium
import pandas as pd

# Sample DataFrame with LOCATION column
data = {
    'LOCATION': ['(40.7128, -74.0060)', '(40.730610, -73.935242)', '(40.751244, -73.980769)']
}
df = pd.DataFrame(data)

# Create a map centered around New York City
nyc_map = folium.Map(location=[40.7128, -74.0060], zoom_start=10)

# Iterate through the 'LOCATION' column and add markers for each location
for index, row in df.iterrows():
    location = row['LOCATION']
    if pd.notna(location):  # Check if location is not missing
        try:
            lat, lon = location.strip('()').split(',')
            lat, lon = float(lat.strip()), float(lon.strip())  # Remove any leading/trailing spaces and convert to float
            folium.Marker([lat, lon]).add_to(nyc_map)
        except ValueError as e:
            print(f"Error parsing location '{location}' at index {index}: {e}")

nyc_map
```
